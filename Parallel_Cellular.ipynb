```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# Load a sample image and convert to grayscale
image = color.rgb2gray(data.astronaut())  # Using skimage's astronaut image
print(f"Original image shape: {image.shape}")

def averaging_filter(img):
    # Pad the image with zeros on all sides to handle edge pixels
    padded_img = np.pad(img, pad_width=1, mode='constant', constant_values=0)

    # Create an output image
    output = np.zeros_like(img)

    # Iterate over each pixel in the image
    # (Vectorized solution will be faster but this is clearer)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # Extract the 3x3 neighborhood
            neighborhood = padded_img[i:i+3, j:j+3]
            # Calculate the average of the neighborhood
            output[i, j] = np.mean(neighborhood)
    return output

# Apply the averaging filter (one iteration)
smoothed_image = averaging_filter(image)

# Plot original and smoothed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Smoothed Image (Averaging Filter)")
plt.imshow(smoothed_image, cmap='gray')
plt.axis('off')

plt.show()
```
